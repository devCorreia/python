**.loc, um método muito importante**
- Pegar linha
- Pegar linhas de acordo com alguma condição 
- Pegar linhas e colunas específicas 
- Pegar 1 valor específico 

```python
# pegar uma linha
 display(vendas_df.loc[6:18])
```

```python
# pegar uma linha com alguma condição
vendas_norte_shopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Shopping Morumbi']
display(vendas_norte_shopping_df)
```

```python
# pegar várias linhas e colunas usando o loc
vendas_norteshopping_df = vendas_df.loc[vendas_df['Data'] == '2019-12-01', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_df)
```

```python
# pegar um valor específico 
print(vendas_df.loc[9, 'Produto'])
```

**Adiconar 1 coluna**



```python
# a partir de uma coluna que existe 
vendas_df["Comissao"] = vendas_df["Valor Final"] * 0.05
display(vendas_df)
```

```python
# criar uma coluna com valor padrão
vendas_df.loc[:, "Imposto"] = 0
display(vendas_df)
```

**Adicionar 1 linha**
- Linhas de um complemento da base de dados

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
display(vendas_dez_df)

#modificando o vendas_df
vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

**Excluir linhas e colunas**

```python
# exclui a coluna Código Venda
vendas_df = vendas_df.drop("Código Venda", axis=1)
display(vendas_df)
```

```python
# exclui a linha 3
vendas_df = vendas_df.drop(3, axis=0)
display(vendas_df)
```

**Valores Vazios**
- Deletar linha/colunas vazias
- Deletar linhas que possuem valores vazios
- Preencher valores vazios (média e último valor)


```python
# deletar linhas e colunas completamente vazias
vendas_df = vendas_df.dropna(how='all', axis=0)
display(vendas_df)
```

```python
# deletar linhas que possuem pelo menos 1 valor sozinho
vendas_df = vendas_df.dropna()
display(vendas_df)
```

```python
# preencher valores vazios
# preencher com a média da coluna
vendas_df["Comissao"] = vendas_df["Comissao"].fillna(vendas_df["Comissao"].mean())
display(vendas_df)
```

```python
# preenche com o último valor acima
vendas_df = vendas_df.ffill()
display(vendas_df)
```

**Calcular indicadores**
- Groupby
- Value Counts

```python
# value counts
transacoes_loja = vendas_df["ID Loja"].value_counts()
display(transacoes_loja)
```

```python
# group by
faturamento_produto = vendas_df[["Produto", "Valor Final",]].groupby("Produto").sum()
display(faturamento_produto)
```

**Mesclar 2 dataframes (Procurar informações de um dataframe em outro)**

```python
gerentes_df = pd.read_excel("Gerentes.xlsx")
# display(gerentes_df)

vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
